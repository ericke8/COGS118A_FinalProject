```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
adult = pd.read_csv("ADULT_EvalResults.csv")
adult.drop('Params', axis = 1)
```

```python
rounded = adult.round(4)
rounded.to_csv('ADULT_Eval_round.csv')
```

```python
partitions = ['0.2', '0.5', '0.8']

modelSet = set(adult['Model'])

acc = adult['Test_Acc'].to_numpy()
acc = acc.reshape(len(modelSet), -1)

df_acc = pd.DataFrame(acc, index = modelSet, columns = partitions)

f1 = adult['Test_F1'].to_numpy()
f1 = f1.reshape(len(modelSet), -1)

df_f1 = pd.DataFrame(f1, index = modelSet, columns = partitions)


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 7))

sns.heatmap(df_acc, annot = True, ax = axes[0], cmap = 'Blues')
axes[0].set_title('ADULT Test Accuracy for Each Partition')

sns.heatmap(df_f1, annot = True, ax = axes[1], cmap='Reds')
axes[1].set_title('ADULT Test F1 for Each Partition')
plt.tight_layout()
plt.savefig('ADULT_hmap.png')
```
